```python
## R Squared is a statistical Measure which tells how well our Regression Model fits the data
```

```python
## It is Calculated by taking the ratio of Sum of squared errors/Total sum of squares and subtracting it by 1
## R-Squared :   1-(Sum of Squared Errors/Total sum of squares)
```

```python
## The sum squared regression is the sum of the residuals squared, and the total sum of squares is the sum of the distance the data is away from the mean all squared. 
## As it is a percentage it will take values between 0 and 1
## R Squared =1 means  	All the variation in the y values is accounted for by the x values
```
